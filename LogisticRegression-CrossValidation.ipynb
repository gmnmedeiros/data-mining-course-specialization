```r
credit_ = credit

dummy.data.frame(credit_, sep="_")

summary(credit_)
```

```r
head(credit_)
```

```r

# Logit com Cross Validation

ctrl <- trainControl(method="repeatedcv", number=20, repeats=5)

fit <- train(X.~., data=credit_,
                  method="glm", 
                  family="binomial",
                  trControl=ctrl, 
                  preProcess = c("center", "scale"))
fit
```

```r
predict_test = predict(fit, newdata=credit_, type="raw")

summary(predict_test)
```

```r
predict_test
```

```r
tail(credit_)
```

```r
pr=prediction(as.numeric(predict_test), as.numeric(credit_$X.)) 
prf=performance(pr, measure="tpr", x.measure="fpr")
plot(prf,colorize=TRUE)
```

```r
# Area under ROC curve 
auc=performance(pr, measure="auc")
auc=auc@y.values[[1]]
auc
```
